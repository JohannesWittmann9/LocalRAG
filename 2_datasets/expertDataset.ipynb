```python
import numpy as np
import json

with open("expert-dataset.json", 'r', encoding="utf8") as file:
        expert_dataset = json.load(file)

relevance_list = []
complexity_list = []

for data in expert_dataset:
        relevance = data["relevance"]
        complexity = data["complexity"]
        relevance_list.append(relevance)
        complexity_list.append(complexity)


print(f"Length of expert dataset: {len(expert_dataset)}")
print(f"Mean complexity: {np.array(complexity_list).mean()}, SD: {np.std(np.array(complexity_list))}")
print(f"Mean relevance: {np.array(relevance_list).mean()}, SD: {np.std(np.array(relevance_list))}")
```
